```python
product_review_counts = df.groupby('ProductId').size().reset_index(name='ReviewCount')
```

```python
product_avg_rating = df.groupby('ProductId')['Score'].mean().reset_index(name='AvgRating')
```

```python
popularity_df = pd.merge(product_review_counts, product_avg_rating, on='ProductId')
```

```python
popular_products = popularity_df.sort_values(by=['ReviewCount', 'AvgRating'], ascending=False)
```

```python
popular_products.head(10)
```

```python
print(" Top 10 Popular Products:")
print(popular_products.head(10))
```

```python
popular_products.to_csv('popular_products.csv', index=False)
```
